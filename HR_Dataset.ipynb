```python
# Load the Excel file
file_name = 'HR_Dataset_Raw.xlsx'
xls = pd.ExcelFile(file_name)

# Display sheet names
print(xls.sheet_names)
```

```python
# Read each sheet into a DataFrame
employees_df = pd.read_excel(xls, sheet_name="Employees_Details")
department_df = pd.read_excel(xls, sheet_name="Department")
attendance_df = pd.read_excel(xls, sheet_name="Attendance")
performance_df = pd.read_excel(xls, sheet_name="Performance_Reviews")
```

```python
import warnings
warnings.filterwarnings('ignore')
# Define columns
num_cols = ['Employee_ID', 'Age', 'Salary', 'Department_ID']

# Fill missing values
employees_df['Employee_ID'].fillna(0, inplace=True)  # Mode is best for categorical IDs
employees_df['Age'].fillna(employees_df['Age'].median(), inplace=True)   # Median is better for Age (avoids outliers)
employees_df['Salary'].fillna(employees_df['Salary'].mean(), inplace=True)  # Mean is useful for Salary
employees_df['Department_ID'].fillna(employees_df['Department_ID'].mode()[0], inplace=True)  # Mode is best for categorical IDs

# Ensure no infinite values
employees_df.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
# Convert to integer
employees_df['Employee_ID'] = employees_df['Employee_ID'].astype(int)
employees_df['Age' ] = employees_df['Age'].astype( int)
employees_df['Salary' ]= employees_df['Salary'].astype( int)
employees_df['Department_ID'] = employees_df['Department_ID'].astype( int)
#employees_df[num_cols] = employees_df[num_cols].astype(int)
```

```python
employees_df.info()
```

```python
# fillna of ojbet type collumn
missing_names = employees_df[employees_df['First_Name'].isnull() | employees_df['Last_Name'].isnull()]
print("No. of missing values: ", len(missing_names))
```

```python
employees_df['Email'].fillna('unknown@email.com', inplace=True)
employees_df['Phone'].fillna(0, inplace=True)
```

```python
# drop null records
employees_df.dropna(subset=['First_Name', 'Last_Name','Joining_Date','Gender'], inplace=True)
```

```python
# Remove invalid emails
import re
employees_df["Email"] = employees_df["Email"].apply(lambda x: x if re.match(r"[^@]+@[^@]+\.[^@]+", str(x)) else np.nan)
employees_df["Email"].fillna("unknown@email.com", inplace=True)

# Format phone numbers (keep only digits and ensure 10 digits)
employees_df["Phone"] = employees_df["Phone"].astype(str).apply(lambda x: re.sub(r'\D', '', x)[:10])
```

```python
missing_attendance_count = attendance_df["Attendance_ID"].isnull().sum()
total_records = len(attendance_df)
print(f"Missing Attendance_IDs: {missing_attendance_count} out of {total_records}")
```

```python
attendance_df["Attendance_ID"] = attendance_df.groupby("Employee_ID")["Attendance_ID"].transform(lambda x: x.fillna(method="ffill").fillna(method="bfill"))
```
